```python
import os
import json
import pandas as pd
import zipfile

# ====== STEP 1: UPLOAD & EXTRACT ZIP FILES ======

# Manually upload ZIP files in Google Colab
from google.colab import files

uploaded = files.upload()  # Upload multiple ZIP files

# Create a folder to store extracted JSON files
os.makedirs("nvd_json_files", exist_ok=True)

# Extract all uploaded ZIP files
for zip_filename in uploaded.keys():
    with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
        zip_ref.extractall("nvd_json_files")
    print(f"✅ Extracted: {zip_filename}")

# ====== STEP 2: LOAD & PROCESS MULTIPLE JSON FILES ======

# Folder where JSON files are stored
nvd_json_folder = "nvd_json_files"

# Initialize empty dictionary for CVE descriptions
cve_dict = {}

# Loop through all JSON files in the extracted folder
for filename in os.listdir(nvd_json_folder):
    if filename.endswith(".json"):
        file_path = os.path.join(nvd_json_folder, filename)
        print(f"Processing {filename}...")

        with open(file_path, "r", encoding="utf-8") as file:
            cve_data = json.load(file)

            # Extract CVE descriptions from each JSON file
            for entry in cve_data.get("CVE_Items", []):  # Ensure we avoid missing keys
                cve_id = entry["cve"]["CVE_data_meta"]["ID"]
                description_data = entry["cve"].get("description", {}).get("description_data", [])

                if description_data:
                    description = description_data[0].get("value", "No description available")
                else:
                    description = "No description available"

                cve_dict[cve_id] = description

print(f"Extracted {len(cve_dict)} CVE descriptions from all JSON files!")

# ====== STEP 3: MATCH DESCRIPTIONS TO YOUR DATASET ======

# Upload your CVE dataset file
uploaded_dataset = files.upload()
dataset_filename = list(uploaded_dataset.keys())[0]

# Load your dataset containing CVE IDs
df = pd.read_csv(dataset_filename)

# Function to get CVE descriptions from local dictionary
def get_cve_description(cve_id):
    return cve_dict.get(cve_id, "Description Not Found")

# Apply function to dataset
df["detailed_description"] = df["cve_id"].apply(get_cve_description)

# ====== STEP 4: SAVE & DOWNLOAD THE UPDATED DATASET ======

# Save the updated dataset
output_filename = "cve_dataset_with_descriptions.csv"
df.to_csv(output_filename, index=False)

# Download the updated dataset
files.download(output_filename)

print("CVE Descriptions Successfully Extracted from ZIP Files!")
```
